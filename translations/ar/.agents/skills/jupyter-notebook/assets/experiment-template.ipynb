```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## الخطة

- الفرضية:
- المتغيرات التي سيتم استكشافها:
- المقاييس التي سيتم تسجيلها:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## النتائج

- الملاحظات الرئيسية:
- المفاجآت أو أوضاع الفشل:
- القرار: الاستمرار، التحول، أم التوقف:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
