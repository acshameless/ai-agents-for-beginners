- সফলতার মানদণ্ড নির্ধারণ করুন।


```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## পরিকল্পনা

- অনুমান:
- স্যুইপ করার পরিবর্তনশীলসমূহ:
- রেকর্ড করার মেট্রিক্স:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## ফলাফল

- মূল পর্যবেক্ষণ:
- অপ্রত্যাশিত ঘটনা বা ব্যর্থতার ধরন:
- সিদ্ধান্ত: চালিয়ে যাওয়া, দিক পরিবর্তন, অথবা বন্ধ করা:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
