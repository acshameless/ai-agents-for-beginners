```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## برنامه

- فرضیه:
- متغیرهایی برای بررسی:
- معیارهایی برای ثبت:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## نتایج

- مشاهدات کلیدی:
- شگفتی‌ها یا حالات شکست:
- تصمیم: ادامه، تغییر مسیر، یا توقف:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
