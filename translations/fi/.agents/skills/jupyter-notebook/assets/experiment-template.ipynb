```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## Suunnitelma

- Hypoteesi:
- Säädettävät muuttujat:
- Tallennettavat mittarit:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## Tulokset

- Keskeiset havainnot:
- Yllätykset tai epäonnistumistilat:
- Päätös: jatkaa, muuttaa suuntaa tai lopettaa:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
