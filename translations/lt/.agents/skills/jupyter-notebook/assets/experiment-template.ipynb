```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## Planas

- Hipotezė:
- Kintamieji, kuriuos keisti:
- Metrikos, kurias užfiksuoti:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## Rezultatai

- Pagrindiniai pastebėjimai:
- Netikėtumai arba gedimų režimai:
- Sprendimas: tęsti, pakeisti kryptį arba sustabdyti:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
