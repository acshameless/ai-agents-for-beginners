## Step 1: Structured рдЖрдЙрдЯрдкреБрдЯрд╕рд╛рдареА Pydantic рдореЙрдбреЗрд▓реНрд╕ рдкрд░рд┐рднрд╛рд╖рд┐рдд рдХрд░рд╛

рд╣реЗ рдореЙрдбреЗрд▓реНрд╕ рдПрдЬрдВрдЯреНрд╕ рдкрд░рдд рдХрд░рддреАрд▓ рдЕрд╢реА **schema** рд╡реНрдпрд╛рдЦреНрдпрд╛ рдХрд░рддрд╛рдд. Pydantic рд╕рд╣ `response_format` рд╡рд╛рдкрд░рд▓реНрдпрд╛рдиреЗ рдЦрд╛рддреНрд░реА рдкрдЯрддреЗ:
- тЬЕ рдкреНрд░рдХрд╛рд░-рд╕реБрд░рдХреНрд╖рд┐рдд рдбреЗрдЯрд╛ рдПрдХреНрд╕реНрдЯреНрд░реЕрдХреНрд╢рди
- тЬЕ рд╕реНрд╡рдпрдВрдЪрд▓рд┐рдд рдкреНрд░рдорд╛рдгреАрдХрд░рдг
- тЬЕ рдореБрдХреНрдд-рдЯреЗрдХреНрд╕реНрдЯ рдкреНрд░рддрд┐рд╕рд╛рджрд╛рдВрдордзреВрди рдХреЛрдгрддреАрд╣реА рдкрд╛рд░реНрд╕рд┐рдВрдЧ рддреНрд░реБрдЯреА рдирд╛рд╣реАрдд
- тЬЕ рдлреАрд▓реНрдбреНрд╕рд╡рд░ рдЖрдзрд╛рд░рд┐рдд рд╕реЛрдкреЗ рд╢рд░реНрддреАрдиреЗ рдорд╛рд░реНрдЧрджрд░реНрд╢рди


````python
class BookingCheckResult(BaseModel):
    """Result from checking hotel availability at a destination."""

    destination: str
    has_availability: bool
    message: str


class AlternativeResult(BaseModel):
    """Suggested alternative destination when no rooms available."""

    alternative_destination: str
    reason: str


class BookingConfirmation(BaseModel):
    """Booking suggestion when rooms are available."""

    destination: str
    action: str
    message: str


print("тЬЕ Pydantic models defined:")
print("   - BookingCheckResult (availability check)")
print("   - AlternativeResult (alternative suggestion)")
print("   - BookingConfirmation (booking confirmation)")
````

## Step 2: рд╣реЙрдЯреЗрд▓ рдмреБрдХрд┐рдВрдЧ рд╕рд╛рдзрди рддрдпрд╛рд░ рдХрд░рд╛

рд╣реЗ рд╕рд╛рдзрди **availability_agent** рд╡рд╛рдкрд░реЗрд▓ рддрдкрд╛рд╕рдгреНрдпрд╛рд╕рд╛рдареА рдХреА рдЦреЛрд▓реА рдЙрдкрд▓рдмреНрдз рдЖрд╣реЗрдд рдХрд╛. рдЖрдореНрд╣реА `@ai_function` рдбреЗрдХреЙрд░реЗрдЯрд░ рд╡рд╛рдкрд░рддреЛ:
- Python рдлрдВрдХреНрд╢рдирд▓рд╛ AI-рдХреЙрд▓ рдХрд░рдгреНрдпрд╛рдпреЛрдЧреНрдп рд╕рд╛рдзрдирд╛рдд рд░реВрдкрд╛рдВрддрд░рд┐рдд рдХрд░рдгреНрдпрд╛рд╕рд╛рдареА
- LLM рд╕рд╛рдареА JSON рд╕реНрдХреАрдорд╛ рдЖрдкреЛрдЖрдк рддрдпрд╛рд░ рдХрд░рдгреНрдпрд╛рд╕рд╛рдареА
- рдкреЕрд░рд╛рдореАрдЯрд░ рд╡реИрдзрддрд╛ рд╣рд╛рддрд╛рд│рдгреНрдпрд╛рд╕рд╛рдареА
- рдПрдЬрдВрдЯреНрд╕рджреНрд╡рд╛рд░реЗ рдЖрдкреЛрдЖрдк рдХреЙрд▓ рд╕рдХреНрд╖рдо рдХрд░рдгреНрдпрд╛рд╕рд╛рдареА

рдпрд╛ рдбреЗрдореЛ рд╕рд╛рдареА:
- **рд╕реНрдЯреЙрдХрд╣реЛрдо, рд╕рд┐рдПрдЯрд▓, рдЯреЛрдХреНрдпреЛ, рд▓рдВрдбрди, рдНрдореНрд╕реНрдЯрд░реНрдбреЕрдо** тЖТ рдЦреЛрд▓реНрдпрд╛ рдЙрдкрд▓рдмреНрдз тЬЕ
- **рдЗрддрд░ рд╕рд░реНрд╡ рд╢рд╣рд░рдВ** тЖТ рдЦреЛрд▓реНрдпрд╛ рдирд╛рд╣реАрдд тЭМ


````python
@ai_function(description="Check hotel room availability for a destination city")
def hotel_booking(destination: Annotated[str, "The destination city to check for hotel rooms"]) -> str:
    """
    Simulates checking hotel room availability.
    
    Returns JSON string with availability status.
    """
    display(
        HTML(f"""
        <div style='padding: 15px; background: #e3f2fd; border-left: 4px solid #2196f3; border-radius: 4px; margin: 10px 0;'>
            <strong>ЁЯФН Tool Invoked:</strong> hotel_booking("{destination}")
        </div>
    """)
    )

    # Simulate availability check
    cities_with_rooms = ["stockholm", "seattle", "tokyo", "london", "amsterdam"]
    has_rooms = destination.lower() in cities_with_rooms

    result = {"has_availability": has_rooms, "destination": destination}

    return json.dumps(result)


print("тЬЕ hotel_booking tool created with @ai_function decorator")
````

## рдЯрдкреНрдкрд╛ 3: рд░реВрдЯрд┐рдВрдЧрд╕рд╛рдареА рд╕реНрдерд┐рддреА рдХрд╛рд░реНрдпреЗ рдкрд░рд┐рднрд╛рд╖рд┐рдд рдХрд░рд╛

рд╣реЗ рдХрд╛рд░реНрдп рдПрдЬрдВрдЯрдЪреНрдпрд╛ рдкреНрд░рддрд┐рд╕рд╛рджрд╛рдЪреА рддрдкрд╛рд╕рдгреА рдХрд░рддрд╛рдд рдЖрдгрд┐ рд╡рд░реНрдХрдлреНрд▓реЛрдордзреНрдпреЗ рдХреЛрдгрддрд╛ рдорд╛рд░реНрдЧ рдШреНрдпрд╛рдпрдЪрд╛ рд╣реЗ рдард░рд╡рддрд╛рдд.

**рдорд╣рддреНрддреНрд╡рд╛рдЪрд╛ рдирдореБрдирд╛:**
1. рддрдкрд╛рд╕рд╛ рдХреА рд╕рдВрджреЗрд╢ `AgentExecutorResponse` рдЖрд╣реЗ рдХрд╛
2. рд╕рдВрд░рдЪрд┐рдд рдЖрдЙрдЯрдкреБрдЯ (Pydantic рдореЙрдбреЗрд▓) рдкрд╛рд░реНрд╕ рдХрд░рд╛
3. рд░реВрдЯрд┐рдВрдЧ рдирд┐рдпрдВрддреНрд░рд┐рдд рдХрд░рдгреНрдпрд╛рд╕рд╛рдареА `True` рдХрд┐рдВрд╡рд╛ `False` рдкрд░рдд рдХрд░рд╛

рд╡рд░реНрдХрдлреНрд▓реЛ рдкреБрдвреАрд▓ рдХреЛрдгрддрд╛ рдХрд╛рд░реНрдпрд╛рдиреНрд╡реЗрд╖рдХ рд╕рдХреНрд░рд┐рдп рдХрд░рд╛рдпрдЪрд╛ рд╣реЗ рдард░рд╡рд┐рдгреНрдпрд╛рд╕рд╛рдареА **рдПрдЬреЗрд╕** рд╡рд░ рдпрд╛ рд╕реНрдерд┐рддреАрдВрдЪреЗ рдореВрд▓реНрдпрд╛рдВрдХрди рдХрд░реЗрд▓.


````python
def has_availability_condition(message: Any) -> bool:
    """
    Condition for routing when hotels ARE available.
    
    Returns True if the destination has hotel rooms.
    """
    if not isinstance(message, AgentExecutorResponse):
        return True  # Default to True if unexpected type

    try:
        result = BookingCheckResult.model_validate_json(message.agent_run_response.text)

        display(
            HTML(f"""
            <div style='padding: 12px; background: #c8e6c9; border-left: 4px solid #4caf50; border-radius: 4px; margin: 10px 0;'>
                <strong>тЬЕ Condition Check:</strong> has_availability = <strong>{result.has_availability}</strong> for {result.destination}
            </div>
        """)
        )

        return result.has_availability
    except Exception as e:
        display(
            HTML(f"""
            <div style='padding: 12px; background: #ffcdd2; border-left: 4px solid #f44336; border-radius: 4px; margin: 10px 0;'>
                <strong>тЪая╕П  Error:</strong> {str(e)}
            </div>
        """)
        )
        return False


def no_availability_condition(message: Any) -> bool:
    """
    Condition for routing when hotels are NOT available.
    
    Returns True if the destination has no hotel rooms.
    """
    if not isinstance(message, AgentExecutorResponse):
        return False

    try:
        result = BookingCheckResult.model_validate_json(message.agent_run_response.text)

        display(
            HTML(f"""
            <div style='padding: 12px; background: #ffecb3; border-left: 4px solid #ff9800; border-radius: 4px; margin: 10px 0;'>
                <strong>тЭМ Condition Check:</strong> no_availability for {result.destination}
            </div>
        """)
        )

        return not result.has_availability
    except Exception as e:
        return False


print("тЬЕ Condition functions defined:")
print("   - has_availability_condition (routes when rooms exist)")
print("   - no_availability_condition (routes when no rooms)")
````

## рдЯрдкреНрдкрд╛ 4: рдХрд╕реНрдЯрдо рдбрд┐рд╕реНрдкреНрд▓реЗ рдПрдХреНрдЭрд┐рдХреНрдпреБрдЯрд░ рддрдпрд╛рд░ рдХрд░рд╛

рдПрдХреНрдЭрд┐рдХреНрдпреБрдЯрд░реНрд╕ рд╣реЗ рдХрд╛рд░реНрдпрдкреНрд░рд╡рд╛рд╣ рдШрдЯрдХ рдЕрд╕рддрд╛рдд рдЬреЗ рд░реБрдкрд╛рдВрддрд░рдгреЗ рдХрд┐рдВрд╡рд╛ рд╕рд╛рдЗрдб рдЗрдлреЗрдХреНрдЯреНрд╕ рдХрд░рддрд╛рдд. рдЕрдВрддрд┐рдо рдкрд░рд┐рдгрд╛рдо рджрд╛рдЦрд╡рд┐рдгреНрдпрд╛рд╕рд╛рдареА рдЖрдореНрд╣реА `@executor` рд╕рдЬрд╛рд╡рдЯрдХрд░реНрддрд╛ рд╡рд╛рдкрд░реВрди рдХрд╕реНрдЯрдо рдПрдХреНрдЭрд┐рдХреНрдпреБрдЯрд░ рддрдпрд╛рд░ рдХрд░рддреЛ.

**рдорд╣рддреНрддреНрд╡рд╛рдЪреЗ рд╕рдВрдХрд▓реНрдкрдирд╛:**
- `@executor(id="...")` - рдПрдЦрд╛рджреНрдпрд╛ рдлрдВрдХреНрд╢рдирд▓рд╛ рдХрд╛рд░реНрдпрдкреНрд░рд╡рд╛рд╣рд╛рдЪрд╛ рдПрдХреНрдЭрд┐рдХреНрдпреБрдЯрд░ рдореНрд╣рдгреВрди рдиреЛрдВрджрдгреА рдХрд░рддреЗ
- `WorkflowContext[Never, str]` - рдЗрдирдкреБрдЯ/рдЖрдЙрдЯрдкреБрдЯрд╕рд╛рдареА рдкреНрд░рдХрд╛рд░ рдорд╛рд░реНрдЧрджрд░реНрд╢рдХ
- `ctx.yield_output(...)` - рдЕрдВрддрд┐рдо рдХрд╛рд░реНрдпрдкреНрд░рд╡рд╛рд╣ рдкрд░рд┐рдгрд╛рдо рдкрд░рдд рдХрд░рддреЛ


````python
@executor(id="display_result")
async def display_result(response: AgentExecutorResponse, ctx: WorkflowContext[Never, str]) -> None:
    """
    Display the final result as workflow output.
    
    This executor receives the final agent response and yields it as the workflow output.
    """
    display(
        HTML("""
        <div style='padding: 15px; background: #f3e5f5; border-left: 4px solid #9c27b0; border-radius: 4px; margin: 10px 0;'>
            <strong>ЁЯУд Display Executor:</strong> Yielding workflow output
        </div>
    """)
    )

    await ctx.yield_output(response.agent_run_response.text)


print("тЬЕ display_result executor created with @executor decorator")
````

````python
# Load environment variables
load_dotenv()

# Check for GitHub Models or OpenAI
chat_client = OpenAIChatClient(base_url=os.environ.get(
    "GITHUB_ENDPOINT"), api_key=os.environ.get("GITHUB_TOKEN"), model_id="gpt-4o")
````

## рдЯрдкреНрдкрд╛ 6: рд╕рдВрд░рдЪрд┐рдд рдЖрдЙрдЯрдкреБрдЯрд╕рд╣ AI рдПрдЬрдВрдЯ рддрдпрд╛рд░ рдХрд░рд╛

рдЖрдкрдг **рддреАрди рд╡рд┐рд╢реЗрд╖ рдПрдЬрдВрдЯреНрд╕** рддрдпрд╛рд░ рдХрд░рддреЛ, рдЬреНрдпрд╛рдкреИрдХреА рдкреНрд░рддреНрдпреЗрдХ `AgentExecutor` рдордзреНрдпреЗ рдЧреБрдВрдбрд╛рд│рд▓реЗрд▓рд╛ рдЖрд╣реЗ:

1. **availability_agent** - рдЯреВрд▓ рд╡рд╛рдкрд░реВрди рд╣реЙрдЯреЗрд▓рдЪреА рдЙрдкрд▓рдмреНрдзрддрд╛ рддрдкрд╛рд╕рдд рдЖрд╣реЗ
2. **alternative_agent** - рдкрд░реНрдпрд╛рдпреА рд╢рд╣рд░реЗ рд╕реБрдЪрд╡рддреЗ (рдЬреЗрд╡реНрд╣рд╛ рдЦреЛрд▓реНрдпрд╛ рдирд╛рд╣реАрдд)
3. **booking_agent** - рдмреБрдХрд┐рдВрдЧрд▓рд╛ рдкреНрд░реЛрддреНрд╕рд╛рд╣рди рджреЗрддреЗ (рдЬреЗрд╡реНрд╣рд╛ рдЦреЛрд▓реНрдпрд╛ рдЙрдкрд▓рдмреНрдз рдЕрд╕рддрд╛рдд)

**рдореБрдЦреНрдп рд╡реИрд╢рд┐рд╖реНрдЯреНрдпреЗ:**
- `tools=[hotel_booking]` - рдПрдЬрдВрдЯрд▓рд╛ рдЯреВрд▓ рдкреНрд░рджрд╛рди рдХрд░рддреЗ
- `response_format=PydanticModel` - рд╕рдВрд░рдЪрд┐рдд JSON рдЖрдЙрдЯрдкреБрдЯрдЪреА рдЦрд╛рддреНрд░реА рдХрд░рддреЗ
- `AgentExecutor(..., id="...")` - рд╡рд░реНрдХрдлреНрд▓реЛ рд╡рд╛рдкрд░рд╛рд╕рд╛рдареА рдПрдЬрдВрдЯрд▓рд╛ рдЧреБрдВрдбрд╛рд│рддреЗ


````python
# Agent 1: Check availability with tool
availability_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a hotel booking assistant that checks room availability. "
            "Use the hotel_booking tool to check if rooms are available at the destination. "
            "Return JSON with fields: destination (string), has_availability (bool), and message (string). "
            "The message should summarize the availability status."
        ),
        tools=[hotel_booking],
        response_format=BookingCheckResult,
    ),
    id="availability_agent",
)

# Agent 2: Suggest alternative (when no rooms)
alternative_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a helpful travel assistant. When a user cannot find hotels in their requested city, "
            "suggest an alternative nearby city that has availability. "
            "Return JSON with fields: alternative_destination (string) and reason (string). "
            "Make your suggestion sound appealing and helpful."
        ),
        response_format=AlternativeResult,
    ),
    id="alternative_agent",
)

# Agent 3: Suggest booking (when rooms available)
booking_agent = AgentExecutor(
    chat_client.create_agent(
        instructions=(
            "You are a booking assistant. The user has found available hotel rooms. "
            "Encourage them to book by highlighting the destination's appeal. "
            "Return JSON with fields: destination (string), action (string), and message (string). "
            "The action should be 'book_now' and message should be encouraging."
        ),
        response_format=BookingConfirmation,
    ),
    id="booking_agent",
)

display(
    HTML("""
    <div style='padding: 15px; background: #e3f2fd; border-left: 4px solid #2196f3; border-radius: 4px; margin: 10px 0;'>
        <strong>тЬЕ Created 3 Agents:</strong>
        <ul style='margin: 10px 0 0 0;'>
            <li><strong>availability_agent</strong> - Checks availability with hotel_booking tool</li>
            <li><strong>alternative_agent</strong> - Suggests alternative cities</li>
            <li><strong>booking_agent</strong> - Encourages booking</li>
        </ul>
    </div>
""")
)
````

## Step 7: рд╕рд╢рд░реНрдд рдХрдбреНрдпрд╛рдВрд╕рд╣ рд╡рд░реНрдХрдлреНрд▓реЛ рддрдпрд╛рд░ рдХрд░рд╛

рдЖрддрд╛ рдЖрдкрдг `WorkflowBuilder` рд╡рд╛рдкрд░реВрди рд╕рд╢рд░реНрдд рдорд╛рд░реНрдЧрдХреНрд░рдордгрд╛рд╕рд╣ рдЧреНрд░рд╛рдл рддрдпрд╛рд░ рдХрд░реВ:

**рд╡рд░реНрдХрдлреНрд▓реЛ рд╕рдВрд░рдЪрдирд╛:**
```

        тЖУ
   Evaluate conditions
        тЖЩ         тЖШ

        тЖУ              тЖУ

        тЖУ              тЖУ
    display_result тЖРтФАтФАтФАтФШ
```

**рдорд╣рддреНрддреНрд╡рд╛рдЪреЗ рдкрджреНрдзрддреА:**
- `.set_start_executor(...)` - рдкреНрд░рд╡реЗрд╢ рдмрд┐рдВрджреВ рд╕реЗрдЯ рдХрд░рддреЗ
