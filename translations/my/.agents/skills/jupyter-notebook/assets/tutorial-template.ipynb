## အစီအစဉ်

1. တပ်ဆင်ခြင်း
2. အနည်းဆုံး လက်တွေ့ အလုပ်လုပ်နိုင်သော ဥပမာ
3. ကွဲပြားမှုများနှင့် အလွဲအမှားများ
4. လေ့ကျင့်ခန်းများ


```python
# Setup cell: keep it short and deterministic
from __future__ import annotations

import math
import random

SEED = 21
random.seed(SEED)
SEED
```

## ခြေလှမ်း ၁ - သေးငယ်တဲ့ ဥပမာနဲ့ စတင်ခြင်း

နောက်ထပ် ဆဲလ်က ဘာလုပ်သလဲ ဆိုတာ ရိုးရှင်းစွာ ရှင်းပြပါ။


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## လေ့ကျင့်ခန်းများ

- အခြားထည့်သွင်းချက်တစ်ခုကို စမ်းကြည့်ပါ.
- ကုဒ်ကို လည်ပတ်ခင် အထွက်ကို ခန့်မှန်းပါ.
- လူများအများဆုံး တွေ့ရသော အမှားတစ်ခုနှင့် ၎င်းကို ပြင်ရန် နည်းလမ်းကို မှတ်သားပါ.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
