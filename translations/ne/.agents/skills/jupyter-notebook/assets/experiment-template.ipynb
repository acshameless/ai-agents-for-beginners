```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## योजना

- परिकल्पना:
- परिवर्तन गर्ने चरहरू:
- रेकर्ड गर्नुपर्ने मेट्रिकहरू:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## नतिजा

- मुख्य अवलोकनहरू:
- आश्चर्य वा असफलता मोडहरू:
- निर्णय: जारी राख्ने, दिशा परिवर्तन गर्ने, वा रोक्ने:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
