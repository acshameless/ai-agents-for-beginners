```python
# Setup cell: keep it short and deterministic
from __future__ import annotations

import math
import random

SEED = 21
random.seed(SEED)
SEED
```

## चरण 1 - सानो उदाहरणबाट सुरु गर्नुहोस्

अर्को सेलले के गर्छ भनेर साधारण भाषामा व्याख्या गर्नुहोस्।


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## अभ्यासहरू

- अर्को इनपुट प्रयास गर्नुहोस्.
- कोड चलाउनु अघि आउटपुट अनुमान गर्नुहोस्.
- एक सामान्य गल्ती र यसलाई कसरी सच्याउने भन्ने कुरालाई नोट गर्नुहोस्.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
