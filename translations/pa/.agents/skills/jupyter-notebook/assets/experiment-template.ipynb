ਉਦੇਸ਼:
- ਉਸ ਸਵਾਲ ਨੂੰ ਦਰਸਾਓ ਜਿਸ ਦਾ ਤੁਸੀਂ ਜਵਾਬ ਦੇਣਾ ਚਾਹੁੰਦੇ ਹੋ.
- ਸਫਲਤਾ ਦੇ ਮਾਪਦੰਡ ਤੈਅ ਕਰੋ.


```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## ਯੋਜਨਾ

- ਪਰਿਕਲਪਨਾ:
- ਬਦਲਣ ਵਾਲੇ ਚਰ:
- ਰਿਕਾਰਡ ਕਰਨ ਲਈ ਮਾਪਦੰਡ:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## ਨਤੀਜੇ

- ਮੁੱਖ ਨਿਰੀਖਣ:
- ਹੈਰਾਨੀਆਂ ਜਾਂ ਨਾਕਾਮੀ ਦੇ ਢੰਗ:
- ਫੈਸਲਾ: ਜਾਰੀ ਰੱਖਣਾ, ਦਿਸ਼ਾ-ਬਦਲਣਾ, ਜਾਂ ਰੁਕਣਾ:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
