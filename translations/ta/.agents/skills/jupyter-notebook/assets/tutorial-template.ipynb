## படி 1 - சிறிய உதாரணத்துடன் தொடங்கு

அடுத்த குழு என்ன செய்கிறது என்பதை எளிய மொழியில் وضح செய்க.


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## பயிற்சிகள்

- ஒரு வேறுபட்ட உள்ளீட்டை முயற்சிக்கவும்.
- குறியீட்டை இயக்குவதற்கு முன் வெளியீட்டை முன்னறிவிக்கவும்.
- ஒரு பொதுவான பிழையை மற்றும் அதை எப்படி சரிசெய்வது என்பதைக் குறிப்பிட்டு வையுங்கள்.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
