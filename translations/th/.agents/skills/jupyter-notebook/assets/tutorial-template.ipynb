```python
# Setup cell: keep it short and deterministic
from __future__ import annotations

import math
import random

SEED = 21
random.seed(SEED)
SEED
```

## ขั้นตอนที่ 1 - เริ่มด้วยตัวอย่างเล็กๆ

อธิบายว่าเซลล์ถัดไปทำอะไรด้วยภาษาธรรมดา.


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## แบบฝึกหัด

- ลองใช้ข้อมูลนำเข้าที่แตกต่าง.
- ทำนายผลลัพธ์ก่อนรันโค้ด.
- จดบันทึกข้อผิดพลาดที่พบบ่อยหนึ่งข้อและวิธีแก้ไข.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
