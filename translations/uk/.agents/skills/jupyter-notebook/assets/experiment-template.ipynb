```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## План

- Гіпотеза:
- Змінні для перебору:
- Метрики для запису:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## Результати

- Ключові спостереження:
- Несподіванки або режими відмов:
- Рішення: продовжувати, змінити напрямок чи зупинитися:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
