## مرحلہ 1 - ایک چھوٹی سی مثال کے ساتھ شروع کریں

اگلے سیل جو کرتا ہے اسے سادہ الفاظ میں بیان کریں۔


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```
