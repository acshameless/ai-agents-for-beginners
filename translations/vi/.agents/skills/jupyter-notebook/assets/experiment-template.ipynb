```python
# Setup: imports and reproducibility
from __future__ import annotations

import random
import statistics

SEED = 7
random.seed(SEED)
SEED
```

## Kế hoạch

- Giả thuyết:
- Các biến để quét:
- Các chỉ số để ghi lại:


```python
# Define parameters and lightweight helpers
sample_size = 20
values = [random.random() for _ in range(sample_size)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## Kết quả

- Những quan sát chính:
- Những điều bất ngờ hoặc chế độ thất bại:
- Quyết định: tiếp tục, chuyển hướng, hay dừng lại:


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
